## Bivariate Analysis

```python
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
df.head()
```

```python
# Scatter plot (Numerical - Numerical) for the relation between two columns
sns.scatterplot(x='math score', y='reading score', hue='gender', style='test preparation course', size='total score', data=df)
plt.show()
```

```python
# Bar plot (Numerical - Categorical)
# Average math scores based on gender and test preparation course
sns.barplot(x='gender', y='math score', hue='test preparation course', data=df)
plt.show()
```

```python
# Box plot (Numerical - Categorical)
# Box plot of math scores based on gender and test preparation course
sns.boxplot(x='gender', y='math score', hue='test preparation course', data=df)
plt.show()
```

```python
# Displot (Distribution Plot)
# Distribution of math scores based on gender
sns.displot(data=df, x='math score', hue='gender', kde=True, multiple="stack")
plt.show()
```

```python
# Heatmap (Categorical - Categorical)
heatmap_data = pd.crosstab(df['gender'], df['test preparation course'])

sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Test Preparation Course by Gender')
plt.show()
```

```python
# Pairplot 
sns.pairplot(df, hue='gender')
plt.show()
```

```python
stockdf = pd.read_csv('./stockDataset.csv')
```

```python
stockdf.head()
```

```python
stockdf['Date'] = pd.to_datetime(stockdf['Date'])

# Line plot for financial data over time
sns.lineplot(x='Date', y='Close', data=stockdf)
plt.title('Stock Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```

```python
# Convert 'Date' to datetime format
stockdf['Date'] = pd.to_datetime(stockdf['Date'])

# Extracting month and year from the 'Date' column
stockdf['Month'] = stockdf['Date'].dt.month
stockdf['Year'] = stockdf['Date'].dt.year

sns.heatmap(stockdf.pivot_table(values='Close',index='Month',columns='Year'))
```

```python
# Creating a pivot table
pivot_table = stockdf.pivot_table(values='Close', index='Month', columns='Year', aggfunc='mean')

# Filling missing values with 0
pivot_table_filled = pivot_table.fillna(0)

# Using sns.clustermap directly on the filled pivot table
sns.clustermap(pivot_table_filled, cmap='YlGnBu', annot=True, fmt=".2f", linewidths=.5)
plt.title('Clustermap of Close Prices Over Months and Years')
plt.show()
```
